##### Exercício 3. Transformação dos dados - Normalização Max-Min

```python
dfMamo_exerc_3 = dfMamo[dfMamo['Idade']!='?']
dfMamo_exerc_3['Idade'] = dfMamo_exerc_3['Idade'].astype(float)
```

```python
def lowest_value(list):
  lowest_value = list[0]
  for value in list:
    if value < lowest_value:
      lowest_value = value
  return lowest_value

def greatest_value(list):
  greatest_value = list[0]
  for value in list:
    if value > greatest_value:
      greatest_value = value
  return greatest_value
```

```python
def max_in_normalize(series):
  min_value = lowest_value(series)
  max_value = greatest_value(series)

  for i, value in series.iteritems():
    new_value = ((value - min_value)/(max_value - min_value))
    series[i] = "{:.2f}".format(new_value)
  
  return series
```

```python
dfMamo_exerc_3['Idade'] = max_in_normalize(dfMamo_exerc_3['Idade'])

dfMamo_exerc_3
```

##### Exercício 4. Redução dos dados - Amostragem Estratificada

```python
sample_size = 100

samples_per_class = []

for severity in dfMamo['Severidade'].unique():
  severity_indices = dfMamo['Severidade'] == severity
  instances = dfMamo[severity_indices]

  proportion_in_original_dataframe = len(instances) / len(dfMamo)
  quantity_in_new_dataframe = round(proportion_in_original_dataframe * sample_size)

  sample = instances.sample(quantity_in_new_dataframe)
  samples_per_class.append(sample)

dfMamo_exerc_4 = pd.concat(samples_per_class)

dfMamo_exerc_4
```
